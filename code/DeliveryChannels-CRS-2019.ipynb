```python
#conversion from constant 2019 million USD to constant 2020 million USD prices


# https://www.census.gov/topics/income-poverty/income/guidance/current-vs-constant-dollars.html
cpi_2019 = 376.5
cpi_2020 = 381.2

dff['Amount_2020_constant_million_USD'] = dff['USD_Disbursement'] * (cpi_2020 / cpi_2019)
```

## Detailed donor - recipient overview by country

```python
recipient_agency_overview = dff.groupby(by=['RecipientName','DonorName', 'AgencyName', 'donor_key' ,'ChannelName', 'ParentChannelCode'])['Amount_2020_constant_million_USD'].sum().reset_index()
recipient_agency_overview.head()
```

## Of all th emoney that goes into a country, what share does go to which channel of delivery?

```python
pcc = pd.read_csv('../data/ParentChannelCodes.csv', delimiter=';')
pcc
```

```python
recipient_agency_overview.head()
```

```python
recipient_agency_overview_m = recipient_agency_overview.merge(pcc, left_on='ParentChannelCode', right_on='ParentChannelCode')
```

```python
def delivery_overview(recipient_country):
        
        country_df = recipient_agency_overview_m[recipient_agency_overview_m['RecipientName']==recipient_country]
        
        total_amount = country_df['Amount_2020_constant_million_USD'].sum()
        
        country_flow = country_df.groupby(by='ChannelName_EN')['Amount_2020_constant_million_USD'].sum().reset_index()
        
        country_flow['share_ODA'] = country_flow['Amount_2020_constant_million_USD']  / total_amount * 100
        
        return country_flow.sort_values(by='share_ODA' ,ascending = False)
```

```python
for country in recipient_agency_overview['RecipientName'].unique():
    print(country)
    country_table = delivery_overview(country)
    display(country_table)
```

```python
def donor_overview(recipient_country):
        
        country_df = recipient_agency_overview[recipient_agency_overview['RecipientName']==recipient_country]
        
        total_amount = country_df['Amount_2020_constant_million_USD'].sum()
        
        country_flow = country_df.groupby(by='donor_key')['Amount_2020_constant_million_USD'].sum().reset_index()
        
        country_flow['share_ODA'] = country_flow['Amount_2020_constant_million_USD']  / total_amount * 100
        
        return country_flow.sort_values(by='share_ODA' ,ascending = False)
```

```python
for country in recipient_agency_overview['RecipientName'].unique():
    print(country)
    country_table = donor_overview(country)
    display(country_table)
```

```python
def detailed_overview(recipient_country):
        
        country_df = recipient_agency_overview[recipient_agency_overview['Recipient Name']==recipient_country]
        
        total_amount = country_df['Amount_2020_constant_million_USD'].sum()
        
        country_flow = country_df.groupby(by=['donor_key','Channel of Delivery Name'])['Amount_2020_constant_million_USD'].sum().reset_index()
        
        country_flow['share_ODA'] = country_flow['Amount_2020_constant_million_USD']  / total_amount * 100
        
        return country_flow.sort_values(by='share_ODA' ,ascending = False)
```

```python
for country in recipient_agency_overview['Recipient Name'].unique():
    print(country)
    country_table = detailed_overview(country)
    display(country_table)
```
